```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10,0.01)
k1 = 1
k2 = 3
plt.plot(x, np.cos(k1*x), label='cos(k1*x)', color='green')
plt.plot(x, np.cos(k2*x), label='cos(k1*x)', color='red', linestyle='dotted')

plt.title('График двух кривых y(x)', family='serif', size=17)
plt.xlabel('Переменная x')
plt.ylabel('Функции y(k,x)=cos(k∙x)')

plt.show()
```
